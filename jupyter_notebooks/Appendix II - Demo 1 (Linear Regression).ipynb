<h2> Use Keras to perform a simple linear regression</h2>

```python
# The modules we're going to use
from __future__ import print_function
from keras.models import Sequential
from keras.layers import Dense

import numpy as np
import matplotlib.pyplot as plt
#np.random.seed(599)

# When you execute a code to plot with a simple SHIFT-ENTER, the plot will be shown directly under the code cell
%matplotlib inline
```

```python
# Create synthetic data
x = np.linspace(-1,1,300)

# Shuffle x's contents
np.random.shuffle(x)

# Create targets y
y = 0.599*x + 2 + np.random.normal(0, 0.08, (300,))

# Split data into training and testing data sets
x_train, y_train = x[:240],y[:240]
x_test, y_test = x[240:],y[240:]

# plot data
plt.scatter(x_train,y_train)
plt.scatter(x_test,y_test, c='r')
plt.show()
```

```python
# build a neural network with only input layer
model = Sequential()
model.add(Dense(input_dim=1, units=1))

# set loss function and optimization method
model.compile(loss='mean_squared_error', optimizer='sgd', )

# training
model.fit(x_train,y_train,batch_size=30, epochs=300, verbose=2)
```

```python
# testing
mse = model.evaluate(x_test, y_test, verbose=2)
print('test mse:', mse)

# display intercept and coefficients
W, b = model.layers[0].get_weights()
print('intercept=',b, 'Weights=', W)

# plotting the prediction
y_pred = model.predict(x_test)
plt.scatter(x_test,y_test,c='r')
plt.plot(x_test, y_pred)
plt.show()
```
